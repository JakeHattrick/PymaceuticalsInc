```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouseFile = "data/mouse_drug_data.csv"
clinicalFile = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mDF = pd.read_csv(mouseFile)
cDF = pd.read_csv(clinicalFile)

# Combine the data into a single dataset
df = pd.merge(mDF,cDF,'outer')

# Display the data table for preview
df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
groupDF = pd.DataFrame(df.groupby(['Drug','Timepoint'],
                                  as_index=False)['Tumor Volume (mm3)'].mean())

# Convert to DataFrame
groupDF['Tumor Volume (mm3)'] = pd.to_numeric(groupDF['Tumor Volume (mm3)'])
# Preview DataFrame
groupDF.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
#group2DF = pd.DataFrame(df.groupby(['Drug','Timepoint'],
#                                  as_index=False)['Tumor Volume (mm3)'].sem())

# Convert to DataFrame
group2DF = pd.DataFrame(df.groupby(['Drug',
                         'Timepoint'])['Tumor Volume (mm3)'].sem()).reset_index()
# Preview DataFrame
group2DF.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
meanDF = groupDF.pivot(index='Timepoint',columns = 'Drug',
                        values = 'Tumor Volume (mm3)')
meanDF.head()
```

```python
# Generate the Plot (with Error Bars)
tdf = meanDF[['Capomulin','Infubinol','Ketapril','Placebo']]
#tdf.head()
# Save the Figure
plt.plot(tdf['Capomulin'],marker = 'x')
plt.plot(tdf['Infubinol'],marker = 'o')
plt.plot(tdf['Ketapril'],marker = '^')
plt.plot(tdf['Placebo'],marker = 's')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid()
plt.legend()
#tdf.plot(grid=True)
plt.title('Tumor Response to Treatment')
plt.savefig('TumorResponseChart.png')
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 

# Convert to DataFrame

# Preview DataFrame
group3DF = pd.DataFrame(df.groupby(['Drug',
                         'Timepoint'])['Metastatic Sites'].mean())
group3DF.reset_index(inplace=True)
group3DF.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

# Convert to DataFrame

# Preview DataFrame

group4DF = pd.DataFrame(df.groupby(['Drug',
                         'Timepoint'])['Metastatic Sites'].sem())
group4DF.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
metadf = group3DF.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')
metadf.head()
```

```python
# Generate the Plot (with Error Bars)

# Save the Figure

# Show the Figure

tdf2 = metadf[['Capomulin','Infubinol','Ketapril','Placebo']]
plt.plot(tdf2['Capomulin'],marker = 'x')
plt.plot(tdf2['Infubinol'],marker = 'o')
plt.plot(tdf2['Ketapril'],marker = '^')
plt.plot(tdf2['Placebo'],marker = 's')
plt.xlabel('Time (Days)')
plt.ylabel('Metastic Spread')
plt.grid()
plt.legend()
plt.title('Metastatic Spread During Treatment')
plt.savefig('MetastaticChart.png')
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

# Convert to DataFrame

# Preview DataFrame

group5DF = pd.DataFrame(df.groupby(['Drug',
                         'Timepoint'])['Mouse ID'].nunique())
group5DF.rename(columns={'Mouse ID':'Mouse Count'},inplace=True)
group5DF.reset_index(inplace=True)
group5DF.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview the Data Frame
countdf = group5DF.pivot(index='Timepoint', columns = 'Drug',values = 'Mouse Count')
countdf.head()
```

```python
# Generate the Plot (Accounting for percentages)

# Save the Figure

# Show the Figure

tdf3 = countdf[['Capomulin','Infubinol','Ketapril','Placebo']]
tdf3 = tdf3/25
plt.plot(tdf3['Capomulin'],marker = 'x')
plt.plot(tdf3['Infubinol'],marker = 'o')
plt.plot(tdf3['Ketapril'],marker = '^')
plt.plot(tdf3['Placebo'],marker = 's')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate')
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()])
plt.grid()
plt.legend()
plt.title('Survival Rate During Treatment')
plt.savefig('SurvivalChart.png')
```
